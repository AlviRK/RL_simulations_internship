```python
#Corridor RL 

import numpy as np
import pandas as pd

np.random.seed(1)

state = 0
temperature = 0.9
discount = 0.9
learning_rate = 0.1
actions = ["left", "right"]
Q = np.zeros((5, 2)) 
history = []
```

```python
def step(state, action):
    if action == "left":
        action = -1
    else:
        action = 1
        
    new_state = state + action
    
    if new_state < 0:
        new_state = max(0, new_state)
    elif new_state > 4:
        new_state = min(4, new_state)

    if new_state == 4:
        reward = 1
    else:
        reward = 0
    return new_state, reward
```

```python
for trial in range (1,1001):
    exp_values = np.exp(Q[state] / temperature)
    probs = exp_values/ np.sum(exp_values)
    action = np.random.choice(["left","right"], p = probs)
    old_state = state
    new_state, reward = step(state,action)
    a_i = actions.index(action)
    target = reward + discount * np.max(Q[new_state])
    prediction_error = target - Q[state, a_i]
    Q[state, a_i] = Q[state, a_i] + learning_rate * prediction_error

    history.append ({
        "time": trial,
        "old_state": old_state,
        "action": action,
        "new_state": new_state,
        "reward": reward,
        "q_left": Q[old_state, 0],
        "q_right": Q[old_state, 1],
        "prediction_error": prediction_error
    })
    
    state = new_state
    if state == 4:
        state = 0
history

df_1 = pd.DataFrame(history)
df_1.iloc[300:401]
```

```python
#Circular Corridor
import numpy as np
np.random.seed(2)
state = 0
actions = ["left","right"]
discount = 0.9
learning_rate = 0.1
Q_values = np.zeros((6,2))
results = []
```

```python
def next_movement(state,action):
    if action == 0:
        action = - 1
    else:
        action = + 1
    
    new_state = state + action
    
    if new_state < 0:
        new_state = 5
    elif new_state > 5:
        new_state = 0

    if new_state == 3:
        reward = 1
    else:
        reward = 0
    
    return(new_state, reward)
        
```

```python
for trial in range(1,131):
    epsilon = 0.3
    if np.random.rand() < epsilon:
        action = np.random.choice([0, 1])
    else:
        action = np.argmax(Q_values[state])
    new_state,reward = next_movement(state,action)
    if action == 0:
        movement = "left"
    else:
        movement = "right"
    
    target = reward + discount * np.max(Q_values[new_state])
    prediction_error = target - Q_values[state,action]
    Q_values[state,action] = Q_values[state,action] + learning_rate * prediction_error

    results.append ({
        "trial": trial,
        "old_state": state,
        "action": movement,
        "new_state": new_state,
        "reward": reward,
        "q_left": Q_values[state, 0],
        "q_right": Q_values[state, 1],
        "prediction_error": prediction_error
    })
    state = new_state

df_circular = pd.DataFrame(results).round(2)
    
pd.set_option("display.max_rows", 100)
pd.set_option("display.max_columns", None)

df_circular.head(100)
    
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 3, figsize=(16, 8), sharey=True)

df_s0 = df_circular[df_circular["old_state"] == 0]
axes[0, 0].plot(df_s0["trial"], df_s0["q_left"], label="Q(0, left)")
axes[0, 0].plot(df_s0["trial"], df_s0["q_right"], label="Q(0, right)")
axes[0, 0].set_title("State 0")
axes[0, 0].set_xlabel("Trial")
axes[0, 0].set_ylabel("Q-value")
axes[0, 0].legend()

df_s1 = df_circular[df_circular["old_state"] == 1]
axes[0, 1].plot(df_s1["trial"], df_s1["q_left"], label="Q(1, left)")
axes[0, 1].plot(df_s1["trial"], df_s1["q_right"], label="Q(1, right)")
axes[0, 1].set_title("State 1")
axes[0, 1].set_xlabel("Trial")
axes[0, 1].legend()

df_s2 = df_circular[df_circular["old_state"] == 2]
axes[0, 2].plot(df_s2["trial"], df_s2["q_left"], label="Q(2, left)")
axes[0, 2].plot(df_s2["trial"], df_s2["q_right"], label="Q(2, right)")
axes[0, 2].set_title("State 2")
axes[0, 2].set_xlabel("Trial")
axes[0, 2].legend()

df_s3 = df_circular[df_circular["old_state"] == 3]
axes[1, 0].plot(df_s3["trial"], df_s3["q_left"], label="Q(3, left)")
axes[1, 0].plot(df_s3["trial"], df_s3["q_right"], label="Q(3, right)")
axes[1, 0].set_title("State 3")
axes[1, 0].set_xlabel("Trial")
axes[1, 0].set_ylabel("Q-value")
axes[1, 0].legend()

df_s4 = df_circular[df_circular["old_state"] == 4]
axes[1, 1].plot(df_s4["trial"], df_s4["q_left"], label="Q(4, left)")
axes[1, 1].plot(df_s4["trial"], df_s4["q_right"], label="Q(4, right)")
axes[1, 1].set_title("State 4")
axes[1, 1].set_xlabel("Trial")
axes[1, 1].legend()

df_s5 = df_circular[df_circular["old_state"] == 5]
axes[1, 2].plot(df_s5["trial"], df_s5["q_left"], label="Q(5, left)")
axes[1, 2].plot(df_s5["trial"], df_s5["q_right"], label="Q(5, right)")
axes[1, 2].set_title("State 5")
axes[1, 2].set_xlabel("Trial")
axes[1, 2].legend()

plt.tight_layout()
plt.show()
```

```python
# 3X3 maze with 9 states
np.random.seed(3)
state = (2,0)
actions = ["up", "down", "left", "right"]
Qs = np.zeros((9,4))
epsilon = 0.3
discount = 0.9
learning_rate = 0.1
history = []
```

```python
def next_step(state, action):
    row, col = state
    
    if action == 0:
        candidate_row = row - 1
        candidate_col = col
    elif action == 1:
        candidate_row = row + 1
        candidate_col = col
    elif action == 2:
        candidate_row = row
        candidate_col = col - 1
    elif action == 3:
        candidate_row = row
        candidate_col = col + 1

    #Moving outside the maze keeps the agent in the same state
    if candidate_row < 0 or candidate_row > 2 or candidate_col < 0 or candidate_col > 2:
        new_state = state
    else:
        new_state = (candidate_row, candidate_col)
    
    #Reward allocation
    if new_state == (0,2):
        reward = 1
    elif new_state == (1,1):
        reward = - 1
    else:
        reward = 0
    
    # If the agent reaches the goal or trap, reset to state 6
    if new_state == (0,2) or new_state == (1,1):
        new_state = (2,0)

    return(new_state,reward)




    
```

```python
for trial in range(1001):
    old_state_id = state[0] * 3 + state[1]
    epsilon = 0.2
    if np.random.rand() < epsilon:
        action = np.random.choice([0,1,2,3])
    else:
        action = np.argmax(Qs[old_state_id])

    if action == 0:
        action_text = "up"
    elif action == 1:
        action_text = "down"
    elif action == 2:
        action_text = "left"
    else:
        action_text = "right"

    new_state,reward = next_step(state,action)
    new_state_id = new_state[0] * 3 + new_state[1]
    
        
    target = reward + discount * np.max(Qs[new_state_id])
    prediction_error = target - Qs[old_state_id, action]
    Qs[old_state_id,action] = Qs[old_state_id, action] + learning_rate * prediction_error
    state = new_state

    history.append ({
        "trial": trial,
        "old_state": old_state_id,
        "action": action_text,
        "new_state": new_state_id,
        "reward": reward,
        "prediction_error": prediction_error,
        "q_up":Qs[old_state_id, 0],
        "q_down":Qs[old_state_id, 1],
        "q_left":Qs[old_state_id, 2],
        "q_right":Qs[old_state_id, 3] 
    })

df_maze = pd.DataFrame(history).round(2)
    
pd.set_option("display.max_rows", 100)
pd.set_option("display.max_columns", None)

df_maze
    
```
